```python
def get_euclidean_distance(a, b):
    x = (b[0] - a[0])**2
    y = (b[1] - a[1])**2
    return (x + y)**(1/2)
```

```python
def get_point_classification(p, bc, gc, oc, pc):
    distances = [
        (get_euclidean_distance(p, bc), "Blue"), (get_euclidean_distance(p, gc), "Green"), 
        (get_euclidean_distance(p, oc), "Orange"), (get_euclidean_distance(p, pc), "Purple")
    ]
    min_distance = distances[0]
    for index in range (1, len(distances)):
        if distances[index][0] < min_distance[0]:
            min_distance = distances[index]
    return min_distance[1]
```

```python
# Generiamo casualmente 128 punti
n_samples = 256
centers = 4
cluster_std = 0.60
random_state = 0
color_points = "White"
file_name = "k_means_data.csv"
points = utils.get_random_points(n_samples, centers, cluster_std, random_state, color_points, file_name)
```

```python
""" Step 1: Generiamo casualmente k centroidi  """
num_bit = 5
min_value = -0.9000
max_value = 0.9000
k = 4
centroids = []
for index in range(k):
    while True:
        x = round(random.uniform(min_value, max_value), 4)
        y = round(random.uniform(min_value, max_value), 4)
        new_centroid = (x, y)
        if new_centroid not in centroids:
            centroids.append(new_centroid)
            break
(bc, gc, oc, pc) = (centroids[0], centroids[1], centroids[2], centroids[3])
```
